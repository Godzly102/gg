```python
%%sql

SELECT 
    [Country Name],
    (MAX(Value) - MIN(Value)) AS Growth
FROM 
    GDP
WHERE 
    Year >= strftime('%Y', 'now') - 20
GROUP BY 
    [Country Name]
ORDER BY 
    Growth DESC;
```

```python
%%sql
SELECT 
    [Country Name],
    ((MAX(Value) - MIN(Value)) / MIN(Value)) * 100 AS Growth_Percentage
FROM 
    GDP
WHERE 
    Year >= strftime('%Y', 'now') - 10
GROUP BY 
    [Country Name]
ORDER BY 
    Growth_Percentage DESC;
```

```python
%%sql
SELECT 
    [Country Name],
    MIN(Year) AS Start_Year,
    MAX(Year) AS Present_Year,
    ((MAX(Value) - MIN(Value)) / MIN(Value)) * 100 AS Growth_Percentage
FROM 
    GDP
WHERE 
    Year >= strftime('%Y', MAX(Year))- 20
GROUP BY 
    [Country Name]
ORDER BY 
    Growth_Percentage DESC;
```

```python
%%sql
SELECT 
    [Country Name],
    MIN(Year) AS Start_Year,
    (SELECT MAX(Year) FROM GDP) AS Present_Year,
    ((MAX(Value) - MIN(Value)) / MIN(Value)) * 100 AS Growth_Percentage
FROM 
    GDP
WHERE 
    Year >= (SELECT MAX(Year) - 20 FROM GDP)
GROUP BY 
    [Country Name]
ORDER BY 
    Growth_Percentage DESC;
```

```python
%%sql
SELECT 
    [Country Name],
    AVG((Value - Prev_Value) / Prev_Value) * 100 AS Average_Annual_Growth
FROM (
    SELECT 
        [Country Name],
        Year,
        Value,
        LAG(Value) OVER (PARTITION BY [Country Name] ORDER BY Year) AS Prev_Value
    FROM 
        GDP
) AS yearly_changes
WHERE 
    Year >= (SELECT MAX(Year) - 20 FROM GDP)
GROUP BY 
    [Country Name]
ORDER BY 
    Average_Annual_Growth DESC;
```

```python
import sqlite3
import pandas as pd

# Connect to the SQLite database
conn = sqlite3.connect('jupyter_sql.db')

# Define the SQL query
sql_query = """
WITH yearly_changes AS (
    SELECT 
        [Country Name],
        Year,
        Value,
        LAG(Value) OVER (PARTITION BY [Country Name] ORDER BY Year) AS Prev_Value
    FROM 
        GDP
)
SELECT 
    [Country Name],
    MIN(Year) AS Start_Year,
    (SELECT MAX(Year) FROM GDP) AS Present_Year,
    AVG((Value - Prev_Value) / Prev_Value) * 100 AS Average_Annual_Growth_Rate,
    CAST(MAX(CASE WHEN Year = (SELECT MAX(Year) FROM GDP) THEN Value END) AS INT) / 1000000000 AS Current_GDP_Billion
FROM 
    yearly_changes
WHERE 
    Year >= (SELECT MAX(Year) - 20 FROM GDP)
    AND [Country Name] IN (
        'United States', 'China', 'Japan', 'Germany', 'India',
        'United Kingdom', 'France', 'Brazil', 'Italy', 'Canada',
        'South Korea', 'Russia', 'Australia', 'Spain', 'Mexico',
        'Indonesia', 'Netherlands', 'Saudi Arabia', 'Turkey', 'Switzerland'
    )
GROUP BY 
    [Country Name]
ORDER BY 
    Average_Annual_Growth_Rate DESC;
"""

# Execute the SQL query and fetch the results
df = pd.read_sql_query(sql_query, conn)

# Close the database connection
conn.close()

# Display the DataFrame
print(df)
```
