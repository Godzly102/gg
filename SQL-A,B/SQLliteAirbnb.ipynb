```python
import pandas as pd 
import sqlite3
df= pd.read_csv('listings.csv')
df


cnn = sqlite3.connect('jupyter_sql.db')

df.to_sql('Air', cnn)
%load_ext sql

%sql sqlite:///jupyter_sql.db

```

```python
%%sql
SELECT id,
       listing_url,
       name,host_location,
       CAST(REPLACE(price, '$', '') AS INTEGER) AS price_integer,
       (CAST(REPLACE(price, '$', '') AS INTEGER)) * availability_365 AS revenue
FROM Air
ORDER BY revenue DESC
LIMIT 20;

```

```python
%%sql
SELECT id,
       listing_url,
       name,host_location,
       CAST(REPLACE(price, '$', '') AS INTEGER) AS price_integer,
       (CAST(REPLACE(price, '$', '') AS INTEGER)) * availability_30 AS revenue_30


FROM Air
ORDER BY revenue_30 DESC
LIMIT 20;
SELECT id,
       listing_url,
       name,
       
       CAST(REPLACE(price, '$', '') AS INTEGER) AS price_integer,
       (CAST(REPLACE(price, '$', '') AS INTEGER)) * availability_30 AS revenue
FROM Air
WHERE availability_365 > 0  -- Exclude rows where availability_365 is 0
ORDER BY revenue DESC
LIMIT 20;
```
